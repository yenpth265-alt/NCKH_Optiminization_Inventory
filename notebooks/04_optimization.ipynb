```python
import pandas as pd
import numpy as np
from budget_constraint import optimize_inventory_knapsack

# 1. Giả sử df của bạn đã được load (Walmart M5)
# df = pd.read_csv('walmart_processed.csv') 

# 2. Thiết lập ngân sách
MY_BUDGET = 100000 

# 3. Gọi hàm tối ưu (đảm bảo file .py đã lưu logic "trừ shortage_cost")
total_val, df_final_plan = optimize_inventory_knapsack(df, MY_BUDGET, scale=1)

# --- 4. HIỂN THỊ KẾT QUẢ CHI TIẾT ---
print(f"{' QUẢN TRỊ TỒN KHO TỐI ƯU ':=^50}")
print(f"💰 Lợi nhuận thực tế (đã trừ phạt): {total_val:,.2f} VNĐ")

# Tính toán thêm thông số để báo cáo NCKH:
total_penalty_if_zero = df['shortage_cost'].sum()
saved_penalty = df_final_plan['shortage_cost'].sum() if not df_final_plan.empty else 0

print(f"⚠️ Tổng mức phạt nếu không nhập hàng: -{total_penalty_if_zero:,.2f} VNĐ")
print(f"✅ Tiền phạt đã 'cứu' được: +{saved_penalty:,.2f} VNĐ")
print("-" * 50)

if not df_final_plan.empty:
    print(f"📦 DANH SÁCH {len(df_final_plan)} SẢN PHẨM NÊN NHẬP:")
    # Hiển thị các cột quan trọng
    cols_to_show = ['item_id', 'demand', 'total_cost', 'net_gain_if_buy', 'shortage_cost']
    display(df_final_plan[cols_to_show])
    
    used_budget = df_final_plan['total_cost'].sum()
    print(f"\n📊 Ngân sách đã sử dụng: {used_budget:,.2f} / {MY_BUDGET:,.2f} ({(used_budget/MY_BUDGET)*100:.1f}%)")
else:
    print("❌ Cảnh báo: Ngân sách hiện tại không đủ để tối ưu hóa bất kỳ sản phẩm nào.")
```
